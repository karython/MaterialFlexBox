## Template Blank

Ao criarmos uma aplicação com o template default, ele já vem com varias dependencias instaladas e configuradas, o que nos permite testar nosso app tanto no celular quanto na web, pois possui o react-dom, expo-web-browser, que são dependencias necessarias para executar na web.

O template blank, é mais enxuto, voutado para quem deseja começar o app do zero. Precisa ter atenção nesse ponto, pois toda dependencia precisará ser instalada. Quando adotamos esse estilo de programação, temos que ficar atentos as versoes de cada biblioteca instalada, pois podem haver incompatibilidade de versões.

A versão atual do Expo SDK é a 55, lançada em 22 de janeiro de 2026, que inclui o React Native 0.83. Já a versão disponivel nas plataformas de aplicativos AppleStore e PlayStore para o Expo GO é do SDK 54, o que gera um erro de incompatibilidade de versão. 

```bash
# baixando versão específica do SDK do Expo
npm install expo@54.0.6

# para template Blank, dependencias da Web
npx expo install react-dom react-native-web

# versao recente 19.2.0, gera incompatibilidade com react-dom 19.1.0
npx expo install react@19.1.0

npx expo install expo-web-browser


```

> Todas essas configurações ficam visíveis no ```pakage.json```
